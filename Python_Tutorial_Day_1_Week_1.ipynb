## Creating Virtual Environment

It is good practice to create a virtual environment and do programming in that. This way you are ensuring that the base installation won't get affected if somethings go wrong.

The <i><b>venv</b></i> module provides support for creating lightweight virtual environments with their own site directories, optionally isolated from system site directories. Each virtual environment has its own Python binary (which matches the version of the binary that was used to create this environment) and can have its own independent set of installed Python packages in its site directories.

<li>Creation of virtual environments is done by executing the command venv in the command prompt:</li>
<code>c:\>c:\Python38\python -m venv c:\path\to\myenv</code>


<li>The command, if run with -h, will show the available options:</li>
<code>venv [-h] [--system-site-packages] [--symlinks | --copies] [--clear]
            [--upgrade] [--without-pip] [--prompt PROMPT] [--upgrade-deps]
            ENV_DIR [ENV_DIR ...]</code>

Once you setup the virtual environment, it is time to activate it.<br>
<ul style="list-style-type:square;">
    <li>Open up <em>cmd</em></li>
    <li>Navigate to the virtual enviroment directory using file browser, e.g. C:\Users\PythonSoftware\envs\IoT_Projects</li>
    <li>Then type <code>Scripts\activate</code></li>
</ul>
To Deactivate, simply type <code>deactivate</code>

<h2>Installing Modules</h2>

<em>pip</em> is the preferred installer program. Starting with Python 3.4, it is included by default with the Python binary installers.
<p>The following command will install the latest version of a module and its dependencies from the Python Packaging Index:</p>
<code>python -m pip install SomePackage</code>

Its also possible to specify an exact or minimum version directly on the command line.<br>
<code>python -m pip install SomePackage==1.0.4    # specific version</code><br>
<code>python -m pip install SomePackage>=1.0.4  # minimum version</code>

Normally, if a suitable module is already installed, attempting to install it again will have no effect. Upgrading existing modules must be requested explicitly:<br>
<code>python -m pip install --upgrade SomePackage</code>
